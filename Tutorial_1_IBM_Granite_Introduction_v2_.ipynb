## 📖 What is IBM Granite?

Before we write any code, let's understand what we're working with.

**IBM Granite** is a family of AI language models developed by IBM Research and released as open source in 2024. "Open source" means the model weights — the mathematical parameters that make the AI work — are freely available for anyone to download and use. This is different from proprietary models like ChatGPT or Claude, where you access the model through a paid API but can't see or control what's inside.

### Why does this matter for faculty and researchers?

| Feature | What it means for you |
|---|---|
| **Open source** | Free to use — no licensing fees, no vendor lock-in, no per-query charges |
| **Multiple sizes** | Small models (2B parameters) that run on free hardware like Colab |
| **Enterprise-grade** | Built by IBM Research for accuracy and reliability, not just demos |
| **Specialized variants** | Separate models optimized for text, code, and structured data |
| **Transparent training data** | IBM documents what data was used to train Granite — important for research ethics and IRB considerations |
| **Local execution** | You can run it on your own infrastructure — data never leaves your control |

That last point is especially important for research. When you use ChatGPT or Claude, your text is sent to a company's servers. With Granite running in Colab, your data stays in your Colab session and is deleted when the session ends. This matters for IRB compliance, FERPA-protected student data, and unpublished research.

### How big is "2 billion parameters"?

Think of parameters as the model's "knowledge capacity." For context:

| Model | Parameters | Analogy |
|---|---|---|
| Granite 3.1 2B (what we'll use) | 2 billion | A knowledgeable graduate student |
| Granite 3.1 8B | 8 billion | An experienced postdoc |
| GPT-4 (estimated) | ~1.8 trillion | An entire research library |

The 2B model is small enough to run for free on Colab's GPU, and capable enough to do genuinely useful work. It won't match GPT-4 on complex reasoning, but for summarization, extraction, and code generation it's surprisingly effective.

### Step 3: Download and load the Granite model

This is the big step. We're downloading the actual AI model from Hugging Face (a platform that hosts open-source models, kind of like GitHub for AI) and loading it into the GPU's memory.

**The model is about 1.5 GB** — think of it as downloading an application. This takes 1–2 minutes on Colab's fast connection.

> **No account needed!** Hugging Face serves popular models without requiring a login. The model downloads directly.

> **SAS/Stata parallel:** This is like `libname mydata '/path/to/data';` in SAS or `use mydata` in Stata — you're telling Python where to find the tool you want to use.

````python
# Step 3: Download and load Granite
#
# What's happening here:
# - AutoTokenizer: Loads the "dictionary" that converts words to numbers
#   (models work with numbers, not text — the tokenizer translates)
# - AutoModelForCausalLM: Loads the actual AI model weights
# - torch_dtype=float16: Uses "half precision" math to save GPU memory
#   (like using FLOAT instead of DOUBLE in SAS — slightly less precise but half the memory)
# - device_map="auto": Automatically puts the model on the GPU

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

MODEL_NAME = "ibm-granite/granite-3.1-2b-instruct"

print(f'Downloading and loading: {MODEL_NAME}')
print(f'Source: https://huggingface.co/{MODEL_NAME}')
print(f'This takes 1-2 minutes on first run...\n')

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto"
)

# Show some info about what we loaded
param_count = sum(p.numel() for p in model.parameters()) / 1e9
device = next(model.parameters()).device

print(f'✅ Model loaded successfully!')
print(f'   Model: {MODEL_NAME}')
print(f'   Parameters: {param_count:.1f} billion')
print(f'   Running on: {device} (GPU)')
print(f'\n   The model is now in GPU memory and ready to generate text.')
````

### Step 4: Create a helper function

Rather than repeating the same code every time we want to ask Granite something, we'll create a reusable function called `ask_granite()`. Think of this as defining a macro in SAS (`%macro ask_granite;`) or a program in Stata (`program define ask_granite`).

**You don't need to understand every line of this code.** What matters is that after running this cell, you can type `ask_granite("your question here")` and get a response.

````python
def ask_granite(prompt, max_new_tokens=512, temperature=0.3):
    """
    Send a prompt to Granite and get a response.

    Parameters:
    -----------
    prompt : str
        Your question or instruction in plain English.

    max_new_tokens : int (default 512)
        Maximum length of the response.
        512 tokens ≈ 1 page of text.
        Increase this if responses are getting cut off.

    temperature : float (default 0.3)
        Controls randomness/creativity.
        0.1 = very focused and deterministic (good for extraction)
        0.3 = balanced (good default)
        0.7 = more creative (good for brainstorming)
        1.0 = maximum creativity (may be less accurate)

    Returns:
    --------
    str : The model's response text
    """
    # Format the prompt using Granite's expected chat template
    messages = [{"role": "user", "content": prompt}]
    formatted = tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )

    # Convert text to numbers (tokenize) and move to GPU
    inputs = tokenizer(formatted, return_tensors="pt").to(model.device)

    # Generate the response
    with torch.no_grad():  # Saves memory — we're not training, just generating
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            temperature=temperature,
            do_sample=True,
            top_p=0.9
        )

    # Decode: convert numbers back to text, skip the original prompt
    response = tokenizer.decode(
        outputs[0][inputs['input_ids'].shape[1]:],
        skip_special_tokens=True
    )
    return response.strip()

# Quick test to make sure everything works
print('Testing the function with a simple question...\n')
test = ask_granite("In one sentence, what is learning analytics?")
print(f'Question: "What is learning analytics?"')
print(f'Granite says: {test}')
print(f'\n✅ Everything is working! You can now use ask_granite() throughout this tutorial.')
````

# 📗 Level 1: Text Summarization

### What is summarization?
Summarization is asking an AI to condense a longer text into a shorter one while preserving the key information. It's the simplest and most common LLM task.

### Why start here?
Summarization has a low barrier to evaluation — you can read the original and the summary and immediately judge whether it's good. This makes it ideal for building intuition about how LLMs work before moving to more complex tasks.

### Practical uses for faculty:
- Condensing research articles for literature reviews
- Creating brief descriptions of course materials for syllabi
- Generating executive summaries of assessment reports for committees
- Preparing meeting briefs from long policy documents

---

### 1.1 Summarize a Research Abstract

Below is a sample abstract about learning analytics in STEM — the kind of text you'd encounter in a literature review. Let's ask Granite to summarize it.

**Notice how the prompt has two parts:**
1. An **instruction** telling the model what to do and how to format the output
2. The **content** to be processed

This two-part structure (instruction + content) is the fundamental pattern for working with LLMs. It's true whether you're using Granite, ChatGPT, Claude, or any other model.

````python
# A sample abstract about learning analytics in STEM higher education
# (This text is based on common findings in the field — not a real paper)

abstract = """
This study examined the relationship between student engagement in an online
learning management system (LMS) and academic performance in undergraduate
STEM courses. Using log data from 573 students across five science courses
(Biology, Physics, Anatomy & Physiology, Forensic Science, and Oceanography),
we analyzed time-on-task, assignment completion rates, and pre/post survey
measures of interest, perceived competence, and value. Results indicated that
time spent in the LMS was a significant predictor of final grade percentage
(β = 0.34, p < .001), even after controlling for prior motivation and course
subject. Students who reported higher pre-course interest spent more time in
the system but did not necessarily earn higher grades, suggesting that
motivation and performance may operate through different mechanisms.
Enrollment status (enrolled vs. withdrawn) was strongly associated with
time-on-task, with withdrawn students logging a median of less than 1 hour.
Implications for early alert systems and targeted interventions are discussed.
"""

# The prompt: instruction + content
prompt = f"""Summarize the following research abstract in 2-3 sentences
that a non-specialist could understand. Focus on the main finding and
its practical implication.

{abstract}"""

# Send to Granite
summary = ask_granite(prompt)

print("=" * 60)
print("ORIGINAL: ~150 words")
print("=" * 60)
print(abstract.strip())
print(f"\n{'=' * 60}")
print(f"GRANITE'S SUMMARY:")
print(f"{'=' * 60}")
print(summary)
print(f"\n💡 The summary captures the key finding (time-on-task predicts")
print(f"   grades) and the nuance (motivation ≠ performance).")
````

### 🔍 What just happened?

Let's break down the process:

1. **You wrote a prompt** with an instruction ("Summarize in 2-3 sentences for a non-specialist") and content (the abstract)
2. **The tokenizer** converted your text into numbers (each word/subword becomes a token ID)
3. **The model** processed those numbers through 2 billion parameters and predicted the most likely next tokens
4. **The tokenizer** converted the output numbers back into readable text

All of that happened in the `ask_granite()` function — you just need to write the prompt.

### The most important principle: **The quality of the output depends on the quality of the prompt.**

A vague prompt like "summarize this" will give a generic summary. A specific prompt like "summarize in 2-3 sentences for a non-specialist, focusing on the main finding" gives you much more control.

### 1.3 Summarize for Different Audiences

A key skill with LLMs is **audience adaptation**. The same content can be rewritten for different readers — a provost cares about different things than a fellow researcher or a first-year student.

This is something that's tedious for humans (rewriting the same thing three times) but trivial for an LLM. Let's see it in action.

````python
# Same abstract, three different audiences
audiences = {
    "University Provost":
        "Summarize this in 2 sentences for a university provost who cares "
        "about student retention rates, resource allocation, and institutional "
        "effectiveness metrics.",

    "Faculty Peer (Researcher)":
        "Summarize the key methodological approach and findings for a fellow "
        "researcher in educational data mining. Use appropriate technical "
        "terminology.",

    "First-Year College Student":
        "Explain what this study found in plain language that a first-year "
        "college student would understand. Avoid jargon and statistical terms."
}

for audience, instruction in audiences.items():
    prompt = f"{instruction}\n\n{abstract}"
    result = ask_granite(prompt, max_new_tokens=200)

    print(f'\n{"="*60}')
    print(f'FOR: {audience}')
    print(f'{"="*60}')
    print(result)

print(f'\n{"="*60}')
print('💡 KEY INSIGHT:')
print('Same information, completely different framing.')
print('The provost version emphasizes retention; the researcher version')
print('emphasizes methods; the student version avoids jargon.')
print('This is the power of prompt engineering.')
````

# 📘 Level 2: Structured Data Extraction

### What is extraction?
Extraction is using an LLM to pull specific, structured information out of unstructured text. You're essentially turning paragraphs into data — like going from a narrative assessment report to a spreadsheet.

### Why this is a game-changer for researchers:
Traditionally, extracting data from text (like coding qualitative responses, or pulling numbers from reports) requires either manual work or complex regular expressions. LLMs can do this with a simple prompt.

### Practical uses:
- Extract key findings, sample sizes, and effect sizes from dozens of papers for a meta-analysis
- Pull course learning outcomes from syllabi into a curriculum map spreadsheet
- Categorize open-ended survey responses into themes
- Convert narrative assessment reports into structured data for accreditation

> **SAS/Stata parallel:** This is like using `INPUT` statements to read messy raw data, except the "data" is English prose instead of delimited fields. You're defining the variables you want, and the LLM figures out how to parse them from the text.

---

### 2.1 Extract Key Information from a Report

Let's extract structured data from a paragraph that describes an early alert system pilot. The paragraph contains numbers, dates, rates, and dollar amounts — all embedded in narrative text.

````python
# A paragraph from a learning analytics report
# Notice: the information we want is scattered throughout the narrative

report_text = """
In Fall 2024, the Department of Biology piloted an early alert system based
on LMS engagement data. Of the 312 students enrolled in Introductory Biology
(BIO 101), 47 were flagged as at-risk during Week 3 based on low login
frequency (fewer than 2 logins per week) and missing assignments. Advisors
contacted 41 of the 47 flagged students (87% contact rate). Of those
contacted, 29 (71%) remained enrolled through the end of the semester, compared
to a historical retention rate of 52% for similar at-risk students. The
intervention cost approximately $3,200 in advisor time (estimated 80 hours at
$40/hour). Course DFW rate decreased from 28% to 21% compared to Fall 2023.
"""

# Ask Granite to extract specific data points
extraction_prompt = f"""Extract the following information from this text.
Format each item as "Label: Value".
If a piece of information is not stated in the text, write "Not stated".

Please extract:
- Course name and number
- Semester and year
- Total enrollment
- Number of at-risk students flagged
- Criteria used for the at-risk flag
- Advisor contact rate
- Retention rate of contacted students
- Historical retention rate (comparison group)
- Total cost of the intervention
- DFW rate before the intervention
- DFW rate after the intervention

Text:
{report_text}"""

extracted = ask_granite(extraction_prompt, max_new_tokens=400)

print('EXTRACTED DATA:')
print('=' * 60)
print(extracted)
print(f'\n💡 Granite pulled 11 specific data points from a single paragraph.')
print(f'   Imagine doing this across 50 program reports for accreditation.')
````

### 2.2 Extract into Machine-Readable Format (JSON)

The extraction above is human-readable, but what if you want to put the data into a spreadsheet or use it in further analysis? We can ask Granite to output **JSON** — a structured data format that Python (and Excel, and everything else) can read.

> **SAS/Stata parallel:** JSON is like creating a SAS dataset with named variables. Instead of `Variable: Value` on screen, you get a structured object with fields you can access programmatically.

````python
# Ask for JSON output — notice how specific the prompt is about the format
json_prompt = f"""Extract information from this text and return it as a JSON object
with exactly these keys and value types:

{{
  "course": "string",
  "semester": "string",
  "total_enrolled": number,
  "at_risk_flagged": number,
  "at_risk_contacted": number,
  "contact_rate_pct": number,
  "retention_rate_pct": number,
  "historical_retention_pct": number,
  "intervention_cost_usd": number,
  "dfw_rate_before_pct": number,
  "dfw_rate_after_pct": number
}}

Return ONLY the JSON object. No other text before or after it.

Text:
{report_text}"""

# Use low temperature for more deterministic/precise output
json_result = ask_granite(json_prompt, max_new_tokens=300, temperature=0.1)
print('Raw output from Granite:')
print(json_result)

# Try to parse it as actual data
import json
try:
    data = json.loads(json_result)
    print(f'\n✅ Successfully parsed as structured data!')
    print(f'\n   Now we can compute with it:')
    if data.get('at_risk_contacted') and data['at_risk_contacted'] > 0:
        cost_per = data.get('intervention_cost_usd', 0) / data['at_risk_contacted']
        print(f'   Cost per contacted student: ${cost_per:.0f}')
    retention_gain = data.get('retention_rate_pct', 0) - data.get('historical_retention_pct', 0)
    print(f'   Retention improvement: +{retention_gain} percentage points')
except json.JSONDecodeError as e:
    print(f'\n⚠️ Could not auto-parse the JSON.')
    print(f'   This happens sometimes with smaller models — the formatting')
    print(f'   may not be perfect. In practice, you would add cleanup code')
    print(f'   or retry the prompt. This is NOT a failure — it\'s the normal')
    print(f'   iteration cycle when working with AI.')
````

### 2.3 Batch Extraction — Process Multiple Texts at Once

This is where LLMs become genuinely practical for research. Instead of manually reading and coding each response, you can process dozens or hundreds of texts automatically.

**Example scenario:** You've collected open-ended course evaluation comments and need to categorize them for your department's annual review.

````python
# Simulated student course evaluation comments
student_comments = [
    "The online labs were really helpful but I wish we had more time for the exams. The TA was great at explaining things.",
    "I couldn't figure out how to navigate the LMS half the time. The content was good but the technology was frustrating.",
    "Best science class I've taken. The instructor made complex topics really understandable. Would definitely recommend to other students.",
    "I dropped the course because I was working 30 hours a week and couldn't keep up with the weekly deadlines. The content itself was fine.",
    "The group projects were a disaster. My teammates never responded to messages. Individual assignments were fine though."
]

print('Analyzing 5 student evaluation comments...\n')
print(f'{"Comment":<10} {"Sentiment":<12} {"Topic":<15} {"Actionable?":<12} {"Key Issue"}')
print('-' * 70)

for i, comment in enumerate(student_comments, 1):
    prompt = f"""Analyze this student course evaluation comment and provide
EXACTLY four items separated by | characters:

Sentiment (positive/negative/mixed) | Main Topic (instruction/technology/workload/assessment/collaboration) | Actionable (yes/no) | Key Issue (5 words max)

Comment: \"{comment}\"

Respond with ONLY the four items separated by |, nothing else."""

    result = ask_granite(prompt, max_new_tokens=60, temperature=0.1)
    # Clean up the result
    result = result.strip().split('\n')[0]  # Take just the first line
    print(f'Comment {i}: {result}')

print(f'\n💡 Imagine running this on 500 evaluation comments.')
print(f'   What takes a committee days of reading could take minutes.')
print(f'   You\'d still want to spot-check the AI\'s categorizations,')
print(f'   but it dramatically reduces the manual workload.')
````

# 📕 Level 3: Code Generation for Data Analysis

### What is code generation?
You describe an analysis in plain English, and the LLM writes working Python code for you. This is particularly powerful for faculty who know *what* they want to do (because they've done it in SAS or Stata) but don't know the Python syntax.

### Why this matters:
- **You don't need to memorize Python syntax.** Describe what you want, get code.
- **Your SAS/Stata knowledge transfers.** You can literally say "do the equivalent of PROC MEANS."
- **You learn Python by example.** Reading generated code with comments is an effective way to pick up a new language.

### Important caveat:
Generated code should always be **reviewed before running on important data.** The 2B model gets most things right, but it can make mistakes — especially with complex statistical procedures. Think of it as a first draft from a research assistant: helpful, but you're still the expert who checks the work.

---

| 3 — Code Generation | Describe analyses in English → get Python | Bridge from SAS/Stata without memorizing syntax |

### Five Principles for Working with LLMs

1. **Be specific in your prompts.** "Summarize in 2 sentences for a provost" beats "summarize this."
2. **Iterate.** The first output rarely needs to be the final output. Refine and re-run.
3. **Verify everything.** LLMs can make confident mistakes. Always check output, especially numbers and code.
4. **Start small.** Test with 1 item before processing 100. Debug on a sample before scaling up.
5. **The prompt is the program.** Investing time in a good prompt pays off more than investing time in code.
